```python
import pandas as pd

# --- 1️⃣ Load your master full copy ---
df = pd.read_csv("Our415_2nd_Cleaned_Copy.csv")

# --- 2️⃣ Quick sanity check ---
print("✅ Loaded full copy for live cleaning")
print("Shape:", df.shape)
print("Columns:", list(df.columns)[:15])
print(df.head(3))
```

```python
# --- Drop non-predictive metadata and descriptive columns ---

meta_cols = [
    "id", "event_name", "event_description", "event_photo", "more_info",
    "site_location_name", "site_address", "site_phone", "site_email",
    "point", "data_as_of", "data_loaded_at"
]

df = df.drop(columns=[c for c in meta_cols if c in df.columns], errors="ignore")

print("✅ Metadata and descriptive columns removed.")
print("Remaining columns:", df.columns.tolist())
print("\nShape after cleanup:", df.shape)
```

```python
# Save your modeling dataset
output_path = "Our415_Modeling_Ready.csv"

df.to_csv(output_path, index=False)
print(f"✅ Saved modeling dataset to '{output_path}'")
print("Shape:", df.shape)
```
