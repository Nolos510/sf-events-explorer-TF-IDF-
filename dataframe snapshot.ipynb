```python
import pandas as pd


# 1) Save your current dataframe snapshot
df = pd.read_csv("Our415_Events_and_Activities_20251008.csv")

# 2) Helpers to make values hashable for nunique()
def make_hashable(x):
    if isinstance(x, list):
        return tuple(make_hashable(i) for i in x)
    if isinstance(x, dict):
        # sort by key for stable representation
        return tuple(sorted((k, make_hashable(v)) for k, v in x.items()))
    if isinstance(x, set):
        return tuple(sorted(make_hashable(i) for i in x))
    return x  # numbers, strings, timestamps, tuples, None/NaN pass through

def safe_nunique(s: pd.Series) -> int:
    # Drop NaNs first to keep dropna=True behavior
    s_no_na = s.dropna()
    # Convert each element to something hashable
    return s_no_na.map(make_hashable).nunique()

# 3) Build the audit summary table
audit_table = pd.DataFrame({
    "column": df.columns,
    "pandas_dtype": [df[c].dtype for c in df.columns],
    "num_unique": [safe_nunique(df[c]) for c in df.columns],
    "non_null_count": [df[c].notna().sum() for c in df.columns],
})

# 4) Save + (optionally) view
audit_table.to_csv("audit_summary_table.csv", index=False)
audit_table
```
